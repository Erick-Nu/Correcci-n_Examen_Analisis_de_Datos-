```python
ruta1 = 'archivo2_4.csv'
archivo1 = pd.read_csv(ruta1,index_col=0)
ruta2 = 'archivo1_3.csv' 
archivo2 = pd.read_csv(ruta2,index_col=0)
BDvehiculos= pd.concat([archivo1,archivo2], ignore_index=True)
# El (ignore_index=True) enumera los indeces nuevamente para que no existan duplicaciones.
BDvehiculos
```

```python

print(f"La fecha más actual es: {BDvehiculos['Año'].min()}")
print()
print(f"La fecha más antigua es: {BDvehiculos['Año'].max()}")
```

```python
new = BDvehiculos[((BDvehiculos['Año'] >= 2020) & (BDvehiculos['Año'] <= 2024))]
print(f"Los autos con fechas actuales son: {new.shape[0]}")
```

```python
new = BDvehiculos[((BDvehiculos['Año'] >= 1990) & (BDvehiculos['Año'] <= 1998))]
print(f"Los autos con fechas antiguas son: {new.shape[0]}")
```

```python
#b.	El promedio de kilometraje de los autos de Quito o Guayaquil.
kilometraje = BDvehiculos[(BDvehiculos['Lugar'] == 'Quito')]
print(f"El Kilometraje de los autos de Quito es: {round(kilometraje['Kilómetraje'].mean(),2)}")
```

```python
#c.	El promedio de kilometraje de los autos de Cuenca.
kilometraje = BDvehiculos[(BDvehiculos['Lugar'] == 'Cuenca')]
kilometraje['Kilómetraje'].mean()
print(f"El Kilometraje de los autos de Cuenca es: {round(kilometraje['Kilómetraje'].mean(),2)}")
```

```python
#e.	Porcentaje de autos nuevos y usados.
total = len(BDvehiculos)
new = BDvehiculos[(BDvehiculos['Categoría'] == 'NEW')]
porcentaje = (len(new) / total) * 100
print(f"El porcentaje de autos new is: {porcentaje:.2f}%")
used= BDvehiculos[(BDvehiculos['Categoría'] == 'USED')]
porcentaje = (len(used) / total) * 100
print(f"El porcentaje de autos used is: {porcentaje:.2f}%")
```

```python
#f.	Las marcas que más se repiten.
marcas= BDvehiculos['Marca'].value_counts()
print("Las marcas más repetidas son:")
marcas.head(10)
```

```python
# i.	Auto más caro y más barato, con sus características.
auto_caro = BDvehiculos['Precio'].max()
print(f"The car more expensive is: {auto_caro} ")
auto_barato = BDvehiculos['Precio'].min()
print(f"The car more cheap is : {auto_barato}")
```

```python
# j.	Auto de Guayaquil que menos kilometraje tiene y más caro es.
carro_guayaquil = BDvehiculos[(BDvehiculos['Lugar'] == 'Guayaquil')]
carro = carro_guayaquil[
    (carro_guayaquil['Kilómetraje'] == carro_guayaquil['Kilómetraje'].min()) &
    (carro_guayaquil['Precio'] == carro_guayaquil['Precio'].max())]
print(f'Los autos de guayaquil que menos kilometraje tinen y más caros son: {carro}')
```
